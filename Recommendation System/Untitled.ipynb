Cosine similarity is a metric used to measure how similar two vectors are based on the angle between them, not their magnitude (length)


Intuition (Simple Idea)

If two vectors point in the same direction → very similar

If they are perpendicular → not similar

If they point in opposite directions → very dissimilar

Cosine Similarity = cos(q) = {A.B/(||A|| ||B||)}

import numpy as np

def cosine_similarity(vec1, vec2):
    """
    Calculate cosine similarity between two vectors
    """
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)

    dot_product = np.dot(vec1, vec2)
    magnitude_1 = np.linalg.norm(vec1)
    magnitude_2 = np.linalg.norm(vec2)

    if magnitude_1 == 0 or magnitude_2 == 0:
        return 0.0   # avoid division by zero

    return dot_product / (magnitude_1 * magnitude_2)

```python
def recommend0(movie):

    movie = movie.strip().lower()
    titles = new_data['title'].str.lower()
    # check if movie exists
    if movie not in titles.values:
        return "Movie not found in data set"
    # get index of the movie
    movie_index = new_data[titles == movie].index[0]
    distances = sim_val[movie_index] --------# get similarity scores
    movies_list = sorted(--------------------# sort by similarity score
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )
    return [
        new_data.iloc[i[0]].title
        for i in movies_list[1:6]
    ]
```

```python
def recommend(movie):
    movie = movie.strip().lower()

    # filter matching rows
    matches = new_data[new_data['title'].str.lower() == movie]

    # check if movie exists
    if matches.empty:
        return "Movie not found"

    # safe index access
    index = matches.index[0]

    distance = sorted(
        list(enumerate(sim_val[index])),
        reverse=True,
        key=lambda vector: vector[1]
    )

    for i in distance[1:6]:  # skip the same movie
        print(new_data.iloc[i[0]].title)

```

```python
recommend("Avatar")
```

```python
from difflib import get_close_matches

def recommend2(movie):
    movie = movie.strip().lower()
    titles = new_data['title'].str.lower()

    matches = get_close_matches(movie, titles, n=1, cutoff=0.6)
    if not matches:
        return "Movie not found"

    movie_index = new_data[titles == matches[0]].index[0]
    distances = sim_val[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )

    return [
        new_data.iloc[i[0]].title
        for i in movies_list[1:6]:
    ]
```
